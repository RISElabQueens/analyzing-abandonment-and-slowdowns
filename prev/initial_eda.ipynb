Creating seperate columns for POPULARITY_1_YEAR	FRESHNESS_numberMissedRelease	FRESHNESS_outdatedTimeInMs

```python
# Function to unescape and parse the JSON string
import json
def parse_freshness(text):
    try:
        # First, unescape the string
        unescaped_text = text.encode('utf-8').decode('unicode_escape')
        # Now, parse the JSON
        data = json.loads(unescaped_text)
        # Extract the 'numberMissedRelease' and 'outdatedTimeInMs'
        number_missed_release = int(data['freshness']['numberMissedRelease'])
        outdated_time_in_ms = int(data['freshness']['outdatedTimeInMs'])
        return pd.Series([number_missed_release, outdated_time_in_ms])
    except (json.JSONDecodeError, TypeError, KeyError):
        return pd.Series([None, None])
    
added_values_refined[['FRESHNESS_numberMissedRelease', 'FRESHNESS_outdatedTimeInMs']] = added_values_refined['FRESHNESS'].apply(parse_freshness)
```

## Feature engineering
Generating release_rhythm, release_count, speed



```python
release['release_timestamp'] = pd.to_datetime(release['release_timestamp'])
#  Merge artifact_release with release to get timestamps for each artifact
artifact_release_info = pd.merge(artifact_release, release[['release_id', 'release_timestamp']], on='release_id', how='left')
#  Calculate the release rhythm for each artifact (list of release timestamps)
rhythm_df = artifact_release_info.groupby('artifact_id')['release_timestamp'].apply(list).reset_index()
rhythm_df.columns = ['artifact_id', 'release_rhythm']
# : Calculate the speed for each artifact
# For each artifact, calculate the number of releases and the total time span (d2 - d1)
artifact_speed = artifact_release_info.groupby('artifact_id').agg(
    release_count=('release_id', 'size'),
    first_release=('release_timestamp', 'min'),
    last_release=('release_timestamp', 'max')
).reset_index()
# Calculate the time interval (d2 - d1) in days
artifact_speed['days_interval'] = (artifact_speed['last_release'] - artifact_speed['first_release']).dt.days

# Avoid division by zero (when there's only one release, set the interval to 1 day)
artifact_speed['days_interval'] = artifact_speed['days_interval'].replace(0, 1)

# Calculate speed (releases per day)
artifact_speed['speed'] = artifact_speed['release_count'] / artifact_speed['days_interval']

#  Merge rhythm and speed into a final dataframe
df_with_features = pd.merge(rhythm_df, artifact_speed[['artifact_id', 'release_count', 'speed']], on='artifact_id')
df_with_features
```

# Is the rhythm of library releases higher than in the past, and how has this rhythm evolved over time?

Further analysis: we can connect POPULARITY_1_YEAR, FRESHNESS_numberMissedRelease, FRESHNESS_outdatedTimeInMs, CVE for further insights

```python
# # Convert release_timestamp to datetime
# release['release_timestamp'] = pd.to_datetime(release['release_timestamp'])

# Step 1: Extract the year from the release timestamps
release['release_year'] = release['release_timestamp'].dt.year

# Step 2: Count the number of releases per year
releases_per_year = release.groupby('release_year').size().reset_index(name='release_count')

# Step 3: Calculate the cumulative number of releases over the years
releases_per_year['cumulative_releases'] = releases_per_year['release_count'].cumsum()
releases_per_year
```

```python
import matplotlib.pyplot as plt
# Step 4: Plot the evolution of release rhythm over time
plt.figure(figsize=(10, 6))
plt.plot(releases_per_year['release_year'], releases_per_year['release_count'], marker='o', linestyle='-', color='b', label='Releases per Year')
plt.title('Evolution of Library Release Rhythm Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid(True)
plt.legend()
plt.show()

# Step 5: Analyze cumulative releases over time (optional)
plt.figure(figsize=(10, 6))
plt.plot(releases_per_year['release_year'], releases_per_year['cumulative_releases'], marker='o', linestyle='-', color='g', label='Cumulative Releases')
plt.title('Cumulative Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Releases')
plt.grid(True)
plt.legend()
plt.show()

# Step 6: Print the release data for analysis
print(releases_per_year)
```

# Do libraries tend to use more dependencies than in the past?
We can further categorize by dependency scopes and analyzse how different scope of dependencies changes i

```python

#  Count the number of dependencies per release
dependencies_per_release = dependency.groupby('release_id').size().reset_index(name='dependency_count')

#  Merge the dependency counts with the release data to get release dates
release_with_deps = pd.merge(release[['release_id', 'release_timestamp']], dependencies_per_release, on='release_id', how='left')

# Fill missing values (releases without dependencies) with 0
release_with_deps['dependency_count'] = release_with_deps['dependency_count'].fillna(0)

#  Extract the year from the release timestamps
release_with_deps['release_year'] = release_with_deps['release_timestamp'].dt.year

#  Calculate the average number of dependencies per release per year
avg_dependencies_per_year = release_with_deps.groupby('release_year')['dependency_count'].mean().reset_index(name='avg_dependencies')

# Step 5: Plot the trend of average dependencies per year
plt.figure(figsize=(10, 6))
plt.plot(avg_dependencies_per_year['release_year'], avg_dependencies_per_year['avg_dependencies'], marker='o', linestyle='-', color='b')
plt.title('Average Number of Dependencies per Release Over Time')
plt.xlabel('Year')
plt.ylabel('Average Number of Dependencies')
plt.grid(True)
plt.show()

# Display the calculated data
print(avg_dependencies_per_year)
```

# Does the emergence of project management methods (e.g., agile methods) have any impact on the release rhythm of libraries?
agile method was introduced in 2001. So I will examine whether libraries started releasing more frequently after this period.

```python
#artifact_release
```

```python
#  Convert release_timestamp to datetime
release['release_timestamp'] = pd.to_datetime(release['release_timestamp'])

#  Define Pre-Agile (before 2001) and Post-Agile (2001 and after)
release['release_year'] = release['release_timestamp'].dt.year
release['agile_period'] = release['release_year'].apply(lambda x: 'Pre-Agile' if x < 2001 else 'Post-Agile')

#  Merge artifact_release with release to get release timestamps for each artifact
artifact_release_info = pd.merge(artifact_release, release[['release_id', 'release_timestamp', 'agile_period']], on='release_id', how='left')

#  Sort the releases by artifact and release timestamp
artifact_release_info = artifact_release_info.sort_values(by=['artifact_id', 'release_timestamp'])

#  Calculate the release intervals (time between consecutive releases) for each artifact
artifact_release_info['release_interval'] = artifact_release_info.groupby('artifact_id')['release_timestamp'].diff().dt.days

# Fill missing values for the first release in each group with 0
artifact_release_info['release_interval'] = artifact_release_info['release_interval'].fillna(0)

#  Calculate the average release interval and release count per period (Pre-Agile vs Post-Agile)
release_intervals_per_period = artifact_release_info.groupby('agile_period').agg(
    avg_release_interval=('release_interval', 'mean'),
    release_count=('release_id', 'size')
).reset_index()

#  Calculate the average speed of releases (releases per year)
release_speed_per_period = release.groupby('agile_period').agg(
    avg_releases_per_year=('release_year', 'size')
).reset_index()

# Display the calculated results
print("\nAverage Release Intervals per Period:")
print(release_intervals_per_period)

print("\nAverage Releases per Year per Period:")
print(release_speed_per_period)
```

# To what extent does the ecosystem contain unmaintained libraries?

Assuming 1 year thresold. 

```python
from datetime import datetime

release['release_timestamp'] = pd.to_datetime(release['release_timestamp'], errors='coerce')

#Merge artifact_release with release to get artifact_id and release_timestamp
merged_data = pd.merge(artifact_release, release[['release_id', 'release_timestamp']], on='release_id', how='left')

# Ensure there are no null values in release_timestamp after the merge
if merged_data['release_timestamp'].isnull().sum() > 0:
    print(f"Warning: {merged_data['release_timestamp'].isnull().sum()} null values in release_timestamp")

# Find the last release date for each library (artifact)
last_release_per_library = merged_data.groupby('artifact_id')['release_timestamp'].max().reset_index()
last_release_per_library.columns = ['artifact_id', 'last_release_date']

# Display a sample of last release dates
print("Sample of last release per library:")
print(last_release_per_library.head())

#  Define the threshold for unmaintained libraries (e.g., no updates in the last 1 years)
threshold_years = 1
current_date = datetime.now()
threshold_date = current_date - pd.DateOffset(years=threshold_years)

print(f"Current date: {current_date}")
print(f"Threshold date (1 years ago): {threshold_date}")

#  Classify libraries as maintained or unmaintained based on the last release date
last_release_per_library['unmaintained'] = last_release_per_library['last_release_date'] < threshold_date

#  Calculate the proportion of unmaintained libraries
total_libraries = len(last_release_per_library)
unmaintained_libraries = last_release_per_library['unmaintained'].sum()
unmaintained_percentage = (unmaintained_libraries / total_libraries) * 100

# Output the result
print(f"Total libraries: {total_libraries}")
print(f"Unmaintained libraries: {unmaintained_libraries}")
print(f"Percentage of unmaintained libraries: {unmaintained_percentage:.2f}%")

print(last_release_per_library[['artifact_id', 'last_release_date', 'unmaintained']].head())
```
